```python
# Import necessary libraries
import pandas as pd
```

## Read Data
Read CSV-files from [output](./output/) directory

```python
# Read CSV files
file1 = "output/summarize-sourcefile-cloud.csv"
df1 = pd.read_csv(file1, delimiter=';')

file2 = "output/summarize-sourcefile-local.csv"
df2 = pd.read_csv(file2, delimiter=';')

file3 = "output/summarize-sourcefile-human.csv"
df3 = pd.read_csv(file3, delimiter=';')

# Append to a combined DataFrame
df_raw = pd.concat([df1, df2, df3], ignore_index=True)

print(f"Raw DataFrame contains {len(df_raw)} rows and {len(df_raw.columns)} columns.")

# DEBUG: Display df_raw as table
df_raw
```

## Filter Data and Reduce Columns

```python
# Filter rows
df = df_raw[df_raw['result_status'] == 'SUCCESS']
df = df[df['label']!='LOCAL']

# Filter out rows with 'grade' not a number
df = df[pd.to_numeric(df['grade'], errors='coerce').notnull()]

# Convert 'grade' to numeric
df['grade'] = pd.to_numeric(df['grade'], errors='coerce')

# Reduce to relevant columns
df = df[['sourcefile', 'label', 'cfg_model', 'grade']]

print(f"Reduced DataFrame to {len(df)} rows and {len(df.columns)} columns.")

# DEBUG: Display df_raw as table, without showing index
df
```

## Generate Scatter-Plot Diagram

```python
# Plotting: Grade per Sourcefile, label datapoints by cfg_model
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 8))  # Increased figure size for better readability

# Generate unique x positions for each sourcefile
sourcefiles = df['sourcefile'].unique()
x_positions = {sourcefile: i for i, sourcefile in enumerate(sourcefiles)}
label_positions = {}

# Plot each row as a scatter point
for _, row in df.iterrows():
    x = x_positions[row['sourcefile']]
    y = row['grade']
    label_model = row['cfg_model']
    label_source = row['label']

    color = 'blue'
    if label_source == 'HUMAN':
        color = 'red'
    elif label_source == 'CLOUD':
        color = 'green'

    plt.scatter(x, y, color=color)

    # Create a key based on x & y for overlapping detection
    key = (x, y)
    count = label_positions.get(key, 0)
    
    # Offset each label vertically by 0.5 per occurrence
    offset = count * 3
    plt.text(x + 0.05 + count * 0.02, y - offset, label_model, fontsize=10, verticalalignment='center')
    
    # Update the counter
    label_positions[key] = count + 1


# Axis labels and title
plt.xticks(
    ticks=list(x_positions.values()),
    labels=list(x_positions.keys()),
    rotation=45,
    ha='right'
)
plt.xlabel("Sourcefile", fontsize=12)
plt.ylabel("Grade", fontsize=12)
plt.title("Grades per Sourcefile grouped by Model", fontsize=14)
plt.ylim(0, 100)    # Force y-axis to range from 0 to 100
plt.gca().set_yscale('linear')  # Explicitly enforce linear scaling

# Improve grid and spacing
plt.grid(True, linestyle='--', alpha=0.5)

# Adjust layout to avoid cut-off labels
plt.subplots_adjust(bottom=0.25)  # Manual bottom margin for x-labels

plt.show()
```

## Generate Boxplot Diagram

```python
# Boxplot visualization of grade per sourcefile
import seaborn as sns

plt.figure(figsize=(16, 8))

# Create a standard boxplot for grade distributions per sourcefile
sns.boxplot(data=df, x="sourcefile", y="grade", color="lightblue")

# Overlay HUMAN datapoints
human_data = df[df['label'] == 'HUMAN']

# Plot HUMAN points as red dots
sns.stripplot(
    data=human_data,
    x="sourcefile",
    y="grade",
    color='red',
    size=8,  # Larger for visibility
    marker='D',  # Diamond marker for clear distinction
    jitter=False
)

# Final plot styling
plt.ylim(0, 100)
plt.xlabel("Sourcefile", fontsize=12)
plt.ylabel("Grade", fontsize=12)
plt.title("Grade Distribution per Sourcefile (Boxplot with HUMAN Results Highlighted)", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45, ha='right')
plt.subplots_adjust(bottom=0.25)

plt.show()
```
